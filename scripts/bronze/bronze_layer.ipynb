#Execute Ingestion

```python
for item in INGESTION_CONFIG:
    print(f'ingesting {item['path'] } -> health.bronze.{item['table']}')
    df = (
        spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(item['path'])
        )
    (
        df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(f"health.bronze.{item['table']}")
    )
    
    
```
