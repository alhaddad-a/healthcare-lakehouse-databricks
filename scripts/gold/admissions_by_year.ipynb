#Read Silver Table

```python
%sql

SELECT *
FROM health.silver.admissions
LIMIT 10;
```

#Business Aggregation

```python
%sql

SELECT
    YEAR(admission_date) AS admission_year,
    COUNT(admission_id) AS total_admissions
FROM health.silver.admissions
GROUP BY
    YEAR(admission_date)
ORDER BY
    admission_year;

```

#Create Gold Table

```python
%sql

CREATE OR REPLACE TABLE health.gold.admissions_by_year AS

SELECT
    YEAR(admission_date) AS admission_year,
    COUNT(admission_id) AS total_admissions
FROM health.silver.admissions
GROUP BY
    YEAR(admission_date);
```

###Gold Table Verification

```python
%sql

SELECT *
FROM health.gold.admissions_by_year
ORDER BY admission_year;
```
