#Read Silver Tables

```python
%sql

SELECT *
FROM health.silver.admissions
LIMIT 10;
```

```python
%sql

SELECT *
FROM health.silver.procedures
LIMIT 10;
```

#Data Integration

```python
%sql

SELECT
    a.admission_id,
    p.procedure_id,
    p.procedure_name,
    p.department,
    p.procedure_cost
FROM health.silver.admissions a
INNER JOIN health.silver.procedures p
    ON a.procedure_id = p.procedure_id;
```

#Business Aggregation

```python
%sql

SELECT
    p.procedure_id,
    p.procedure_name,
    p.department,
    COUNT(a.admission_id) AS times_performed,
    p.procedure_cost
FROM health.silver.admissions a
INNER JOIN health.silver.procedures p
    ON a.procedure_id = p.procedure_id
GROUP BY
    p.procedure_id,
    p.procedure_name,
    p.department,
    p.procedure_cost
ORDER BY
    times_performed DESC;
```

#Create Gold Table

```python
%sql

CREATE OR REPLACE TABLE health.gold.procedure_performance AS

SELECT
    p.procedure_id,
    p.procedure_name,
    p.department,
    COUNT(a.admission_id) AS times_performed,
    p.procedure_cost
FROM health.silver.admissions a
INNER JOIN health.silver.procedures p
    ON a.procedure_id = p.procedure_id
GROUP BY
    p.procedure_id,
    p.procedure_name,
    p.department,
    p.procedure_cost;
```

###Gold Table Verification

```python
%sql

SELECT *
FROM health.gold.procedure_performance
ORDER BY times_performed DESC;
```
