#Data Integration

```python
%sql

SELECT
    b.admission_id,
    b.total_cost,
    a.hospital_id,
    h.hospital_name,
    h.state
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.hospitals h
    ON a.hospital_id = h.hospital_id;
```

###Business Aggregation

```python
%sql

SELECT
    h.hospital_id,
    h.hospital_name,
    h.state,
    SUM(b.total_cost) AS total_revenue
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.hospitals h
    ON a.hospital_id = h.hospital_id
GROUP BY
    h.hospital_id,
    h.hospital_name,
    h.state
ORDER BY total_revenue DESC;
```

#Create Gold Table

```python
%sql

CREATE OR REPLACE TABLE health.gold.revenue_by_hospital AS

SELECT
    h.hospital_id,
    h.hospital_name,
    h.state,
    SUM(b.total_cost) AS total_revenue
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.hospitals h
    ON a.hospital_id = h.hospital_id
GROUP BY
    h.hospital_id,
    h.hospital_name,
    h.state;
```

###Gold Table Verification

```python
%sql

SELECT *
FROM health.gold.revenue_by_hospital
ORDER BY total_revenue DESC;
```
