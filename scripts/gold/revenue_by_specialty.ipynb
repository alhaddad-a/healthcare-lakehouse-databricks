#Data Integration

```python
%sql

SELECT
    b.admission_id,
    b.total_cost,
    d.doctor_id,
    d.specialty
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.doctors d
    ON a.doctor_id = d.doctor_id;
```

#Business Aggregation

```python
%sql

SELECT
    d.specialty,
    SUM(b.total_cost) AS total_revenue
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.doctors d
    ON a.doctor_id = d.doctor_id
GROUP BY
    d.specialty
ORDER BY
    total_revenue DESC;
```

#Create Gold Table

```python
%sql

CREATE OR REPLACE TABLE health.gold.revenue_by_specialty AS

SELECT
    d.specialty,
    SUM(b.total_cost) AS total_revenue
FROM health.silver.billing b
INNER JOIN health.silver.admissions a
    ON b.admission_id = a.admission_id
INNER JOIN health.silver.doctors d
    ON a.doctor_id = d.doctor_id
GROUP BY
    d.specialty;
```

#Gold Table Verification

```python
%sql

SELECT *
FROM health.gold.revenue_by_specialty
ORDER BY total_revenue DESC;
```
