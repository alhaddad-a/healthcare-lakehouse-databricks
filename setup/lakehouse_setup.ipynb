###Select Catalog

```python
%sql

USE CATALOG health;
```

###Create Lakehouse Schemas

```python
%sql

CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Bronze Layer: Raw healthcare data ingested from source CSV files.';

CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'Silver Layer: Cleaned, validated, and standardized healthcare data.';

CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'Gold Layer: Business-ready healthcare KPIs and analytical datasets.';
```

###Create Volume

```python
%sql

CREATE VOLUME IF NOT EXISTS health.bronze.raw_sources
COMMENT 'Stores raw healthcare CSV source files for the Bronze Layer.';
```

###Verify Schemas


```python
%sql

SHOW SCHEMAS;
```

###Verify Volume

```python
%sql

SHOW VOLUMES IN health.bronze;
```
